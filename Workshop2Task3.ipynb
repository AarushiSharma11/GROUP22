```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/workshop2/medical_students_dataset.csv")
```

```python
# dsiplay the first few rows of the dataframe
print("dataset loaded succesfully : ")
print(df.head())
```

```python
# display basic info about the dataframe
print('dataframe information : ')
print(df.info())
```

```python
# check for columns with missing values
missing_values = df.isnull().sum()
print("columns with missing values : ")
print(missing_values[missing_values > 0])
```

```python

# fill missing values based on the nature of the data
# example assumptions:
# - 'Age' column: fill missing values with the median (continuous numeric data).
# - 'Grade' column: fill missing values with the mode (categorical data).
# - 'Scores' column: fill missing values with the mean (continuous numeric data).

if 'Age' in df.columns:
    df['Age'].fillna(df['Age'].median(), inplace=True)
    print("\nFilled missing 'Age' with Median Value.")

if 'Grade' in df.columns:
    df['Grade'].fillna(df['Grade'].mode()[0], inplace=True)
    print("Filled missing 'Grade' with Mode Value.")

if 'Scores' in df.columns:
    df['Scores'].fillna(df['Scores'].mean(), inplace=True)
    print("Filled missing 'Scores' with Mean Value.")
```

```python
# check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"number of duplicated rows : {duplicate_count}")
```

```python
# drop duplicate rows
if duplicate_count > 0:
  df = df.drop_duplicates()
  print("updated dataframe : ")
  print(df)
```
